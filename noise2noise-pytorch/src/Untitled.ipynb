```python
img_list2=[]

img = cv2.imread('/hdd1/PM/Self2Self/noise2noise/noise2noise-pytorch/data/new_ictc/Fake/Ecoflex 00-50/004_0_0.png', 0)
img2 = Image.fromarray(img) 
img_list2.append(img2)
```

```python
import torchvision.transforms.functional as tvF
```

```python
crop_size=128
```

```python
w, h = img_list2[0].size
assert w >= crop_size and h >= crop_size, \
    f'Error: Crop size: {self.crop_size}, Image size: ({w}, {h})'
cropped_imgs = []


for img in img_list2:
    if min(w, h) < crop_size:
        img = tvF.resize(img, (crop_size, crop_size))
    cropped_imgs.append(tvF.crop(img, 0, 581-crop_size, crop_size, crop_size))
```
